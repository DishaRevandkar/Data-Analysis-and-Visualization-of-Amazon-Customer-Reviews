Saving the object reviews of a particular object into a list in order to check the most frequent words among all the available reviews of a product.

```python
# Installing the word cloud package
# Generating the function to determine the significant words and generate a dictionary containing the frequency of those words.

# !pip install wordcloud

import wordcloud
from IPython.display import display

def word_freq(words_list):
    punctuations = '!#$%&"("")"*+, -./:;<=>?@[\]^_`{|}~'
    interesting_words = ["good","great","excellent","useful","easy","efficient","inefficient","useless","crap","wonders","quality",
                        "nice","recommended","perfect","bad","waste","poor","works","delicate","expensive","boring"]
    
    wordlist = []
    
    for each_word in review_list:
        words = each_word.split(" ")
        for word in words:
            if word in interesting_words:
                wordlist.append(word)

    word_dict = {}

    for word in wordlist:
        if word not in word_dict.keys():
            word_dict[word] = wordlist.count(word)
    
    cloud = wordcloud.WordCloud(mode='RGBA',
               background_color='rgba(255, 200, 255, 255)')
    cloud.generate_from_frequencies(word_dict)
    return cloud.to_array()

my_image = word_freq(review_list)
plt.imshow(my_image,interpolation = 'nearest')
plt.title('Most frequenct reviews for yoomall 5M Antenna WIFI',fontsize = 13)
plt.axis('off')
plt.show()
```
