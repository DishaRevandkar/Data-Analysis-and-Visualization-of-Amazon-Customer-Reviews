```python
#Importing csv file to pandas DataFrame
df = pd.read_csv('appl_2_amazon_electronics.csv')
```

```python
df.head(5)
```

```python
#Identifying the Top 10 Products based on product_id and plotting a bar chart for the same.
Product_ID = df["product_id"].value_counts()
plt.figure(figsize = (8,6))
Product_ID[:10].plot(kind='bar')
plt.xlabel('Product ID')
plt.ylabel('Total Number of Reviews Received by the Product')
plt.title("Number of Reviews for Top 10 Products in Electronics Category")
```

```python
#Identifying the Top 10 Products based on product_id and plotting a pie chart for the same.
Product_ID = df["product_id"].value_counts()
plt.figure(figsize=(10,8))
Product_ID[:10].plot(kind='pie', autopct='%1.0f%%')
plt.title("Pie Chart Distribution for Reviews of Top 10 Products in Electronics Category")
```

```python
#Identifying the Bottom 10 Products based on product_id and plotting a bar chart for the same.
Product_ID = df["product_id"].value_counts()
plt.figure(figsize = (8,6))
Product_ID[-10:].plot(kind='bar')
plt.xlabel('Product ID')
plt.ylabel('Total Number of Reviews Received by the Product')
plt.title("Number of Reviews for Bottom 10 Products in Electronics Category")
```

```python
#Identifying the Bottom 10 Products based on product_id and plotting a pie chart for the same.
Product_ID = df["product_id"].value_counts()
plt.figure(figsize=(10,8))
Product_ID[-10:].plot(kind='pie', autopct='%1.0f%%')
plt.title("Pie Chart Distribution for Reviews of Bottom 10 Products in Electronics Category")
```
